# Web Scraping

1. Requests - used to download the webpage i.e. HTML
2. BeautifulSoup- Parses and extracts data from HTML (generally useful for static HTML)
3. Selenium- Automates the browser to load dynamic JS pages

```python
import requests as r
```

```python
url="https://books.toscrape.com/catalogue/the-dirty-little-secrets-of-getting-your-dream-job_994/index.html"
response=r.get(url)

## we send request generally using try,except block 
```

```python
response.status_code
```

```python
response
```

```python
response.content
```

```python
if response.status_code==200:
    with open("scraped_data","w",encoding="utf-8")as f:
        f.write(response.text)
else:
    print("Unable to fetch the HTML content of url")
```

```python
from bs4 import BeautifulSoup

##  this creates a parse tree from data
with open("scraped_data","r",encoding="utf-8")as f:
    html_content=f.read()

soup=BeautifulSoup(html_content,"lxml")
```

```python
soup
```

```python
# find() -> finds the first occurence of the searched tag
#find_all

#soup.find("h2")
all_h2=soup.find_all("h2")
```

```python
for h2 in all_h2:
    print(h2.get_text())
```
