## Matplotlib

As it is known that a picture is equal to thousand words . So it is helpful in finding out the patterns in data. Such as it becomes very easy for us to find outliers. We can also compare the working of handling outliers i.e before and after handling the outliers or missing data.It becomes easier for other non-tech persons to understand the insights of data

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[5,4,2,1,3]
plt.plot(x,y)
plt.show()
```

```python
import pandas as pd
df=pd.read_csv("sample.csv",encoding="utf-8")
```
