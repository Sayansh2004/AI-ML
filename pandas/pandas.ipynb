# Two types of data structure

1. Series - 1-D arrays. (stores homogenous data only).
2. DataFrame -  combination of series.

```python
info={
    "name":"sayansh",
   " marks":[99,95]
}
df=pd.DataFrame(info)
df
```

```python
s=pd.Series([1,2,3,4,5])
s
```

Accessing in series can be done either via indexes or via labels


```python
ser=pd.Series([1,2,3,4,5],index=["Adam","Bob","eve","ayush","akansha"])

# Both will give same output

print(ser[0])
print(ser["Adam"])

```
