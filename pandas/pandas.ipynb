## Feature Engineering

1. Apply Function

```python
df=pd.read_csv("raw_data.csv")

df2=df.copy()
```

```python
df2["tax"]=df2["income"].apply(lambda x:"20%" if x>=50000 else "10%" ) # we created or derived a new feature from existing feature
df2
```

```python
gender_map={"Male":"M","Female":"F","Unknown":"U"}
df2["gender"]=df2["gender"].map(gender_map)
df2

```

```python
df2["country"]=df2["country"].replace("USA","US")
df2
```

```python
df2.columns=["Id","Name","Age","Country","Gender","Income","Tax"]

df2.rename(columns={"Income":"Salary"})
```

```python
df2.sort_values(["Income","Age"],ascending=False)
```

```python
df2["Rannking"]=df2["Income"].rank() # we can use method=dense/min based on the type of data we wanted if we not want any decimal values
df2
```

```python
# shifting the Id column at the end

new_col_order=[col for col in df2.columns if col!="Id"]+["Id"]
df2[new_col_order]
```

# Melt and Pivot

```python
# Melt-> wide format to long format
# Pivot-> Long format to wide format 

# We do this type of convertion because normalisation becomes helpful

# melt(id_vars)
# id_vars=list which we wanted to keep
# value_vars=the list which we wanted to unpivot 
# var_name=the name of new column 
# value_name=value by default


df.melt(id_vars=[],value_vars=[],var_name="",value_name="")
```
